## II) Cleaning / Preprocessing Data:

* While no missing data was detected in the dataset, we identified that certain features such as premium price, coverage limit, and type of policies are presented as lists. To extract valuable insights from these features, we will perform the following preprocessing steps:

  * Extract individual policies from the lists.
  * Calculate the mean and standard deviation of the premium price and coverage limit features for each policy.

* These preprocessing steps will allow us to better understand the distribution and variability of premium prices and coverage limits across different policy types.

```python
# part where we transform the 'Types of Policies Offered' Feature into categories
def clear_policy_column(policy_str):
    new_liste=[]
    policy_list=policy_str.split(',')
    for i in range(len(policy_list)):
        if policy_list[i][0]==' ':
            new_liste.append(policy_list[i][1:].lower())
        else:
            new_liste.append(policy_list[i].lower())
    return  list(set(new_liste))       
def process(row,policy_list):

    policy_str_liste=row['Types of Policies Offered']
    for policy in policy_str_liste:
        if policy.lower() in policy_list :
            row[policy.lower()]=1
    return row
df['Types of Policies Offered']=df['Types of Policies Offered'].apply(lambda x : clear_policy_column(x))

s=[]
for i in df['Types of Policies Offered']:
    s=s+i
sa=list(set(s)) 
print(sa)
df1=pd.DataFrame(np.zeros((10000,len(sa))),columns=sa)
df2=pd.concat([df,df1], axis=1)
df2=df2.apply(lambda row : process(row,sa),axis=1)
```

```python
# Tranform the premium_rate and coverage_Limits into independant column (calculating the mean)
def mean_column(string):
    liste=string.split(',')
    return sum(list(map(float,liste)))/len(liste)
df2['premium_rate_mean']=df2['Premium Rates'].apply(lambda sr : mean_column(sr)) 
df2['coverage_Limits_mean']=df2['Coverage Limits'].apply(lambda sr : mean_column(sr)) 
```

```python
# Policy , 
policy_liste=['valuable cargo insurance','open cargo policy','warehouse-to-warehouse cargo insurance','named perils cargo insurance','all-risk cargo insurance','voyage policy']
dic={}
for policy in policy_liste:
    dic[policy]=df2[df2[policy]==1].filter(['premium_rate_mean','coverage_Limits_mean'])
```

```python
for feature in ['premium_rate_mean','coverage_Limits_mean']:
    plt.figure(figsize=(15,10))

    for key,item in dic.items() :

        sns.distplot(item[feature],label=key,hist=False)
        plt.legend()
    plt.show()
```

```python
def extract_data_column(string,column_liste):
    liste=string.split(',')
    liste=list(map(float,liste))
    liste_re=[max(liste),min(liste),sum(list(map(float,liste)))/len(liste),len(liste)]
    return pd.Series(liste_re, index=column_liste)

def extract_step(string):
    liste=string.split(',')
    liste=list(map(float,liste))
    if len(liste)==1:
        return 0
    else:
        somme=0
        liste.sort()
        for i in range(1,len(liste)):
            somme=(liste[i]-liste[i-1])+somme
        return somme/len(liste)-1 
# PRice Trait
def extract_data_column_2(row1):
    liste_premuim=sorted(list(map(float,row1['Coverage Limits'].split(','))))
    liste_coverage=sorted(list(map(float,row1['Premium Rates'].split(','))))
    liste_final=[]
    for i,j in zip(liste_coverage,liste_premuim):
        liste_final.append(i/j)
    print(liste_final)    
    return liste_final

def extract_data_column_3(row1,column_liste):
    liste_coverage=sorted(list(map(float,row1['Coverage Limits'].split(','))))
    liste_premuim=sorted(list(map(float,row1['Premium Rates'].split(','))))
    liste_final=[]
    for i,j in zip(liste_coverage,liste_premuim):
        liste_final.append(i/j)
    np_liste=np.array(liste_final)    
    return pd.Series([np.median(np.array(liste_premuim)),np.median(np.array(liste_coverage)),np_liste.mean(), np_liste.std(),np.median(np_liste),len(liste_final),np.percentile(np.array(liste_coverage),75),np.percentile(np.array(liste_premuim),75)], index=column_liste)    
```

```python
ratio_column=['premium_median','coverage_limit_median','price_ratio_mean','price_ratio_std','price_ratio_median','price_ratio_num','75_percentile_coverage','75_percentile_premuim']
data4=df2.apply(lambda row : extract_data_column_3(row,ratio_column),axis=1) 
data4
```

## correlations between two variables

```python
def filter_correlation(df,corr_max):
    num_col=df.select_dtypes(exclude=['object'])
    print('**************** Correlation Filtered *******************')
    liste_features=[]
    for num_1 in num_col:
        
        for num_2 in num_col:
            corr=df[num_1].corr(df[num_2])
            if abs(corr)>corr_max and num_1!= num_2:
                liste_features.append(num_1)
                print(num_1 ,' / ' ,num_2,'=',corr)
    return liste_features
liste_features=filter_correlation(df2,0.5)
```

```python
# Correlation matrix using only corr higher than 0.5

plt.figure(figsize=(30,14))
sns.heatmap(df2.filter(set(liste_features)).corr(),annot=True,square=True)
plt.title('Correlation between Variables',size=8)
plt.xticks(size=13)
plt.yticks(size=13)
plt.show()
```

## VIF (Variance Inflation Factor) Feature Selection:


```python
df3=pd.read_excel("insurance_compagies_4.xlsx").drop(['Agency Name','Location'],axis=1)
from sklearn.preprocessing import LabelEncoder
import pandas as pd
encoder = LabelEncoder()  


categorical_features = df3.select_dtypes(include=['object']) 
numerical_features = df3.select_dtypes(exclude=['object'])  
for column in categorical_features.columns:
    
    df3[column] = encoder.fit_transform(df3[column])
```

```python
feature_names = df3.columns
from statsmodels.stats.outliers_influence import variance_inflation_factor 

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(df3, i) for i in range(df3.shape[1])]
vif["features"] = feature_names


print('*******************')
print('VIF Score higher than 10')
print(vif[vif["VIF Factor"]>10])
```

### After conducting variance inflation factor (VIF) analysis for feature selection, we observed that due to the large number of features, there was a risk of losing important data and variables for our clustering purpose. To address this, we decided to perform feature engineering by selecting certain features and applying clustering only on those selected features. The selected features include Financial Solidity, Financial Dynamics, Price Score, Comfort Score, and Service Score. Each of these features represents a clustering approach to a group of existing features in our dataset, allowing us to capture essential information while reducing the dimensionality of the data."

## Choosing an insurance company


#### There are many insurance companies, so distinguish between them can be a challenge. Here are the main points to keep in mind when selecting an insurance company:

### 1. Licensing
Not every company is licensed to operate in each state. As a general rule, you should buy from a company licensed in your state, because then can you rely on your state insurance department to help if there’s a problem. To find out which companies are licensed in your state, contact the state insurance department.

### 2. Price
Many companies sell insurance policies and prices vary greatly from one to another, so it really pays to shop around. Get at least three price quotes from companies, agents and from the Internet. Your state insurance department may publish a guide that shows what insurers charge for different policies in various parts of your state.

### 3. Financial Solidity
You buy insurance to protect you financially and provide peace of mind. Select a company that is likely to be financially sound for many years, by using ratings from independent rating agencies.

###  4. Service
Your insurance company and its representatives should answer your questions and handle your claims fairly, efficiently and quickly. You can get a feel for whether this is the case by talking to other customers who have used a particular company or agent. You may also want to check a national claims database to see what complaint information it has on a company. Also, your state insurance department will be able to tell you if the insurance company you are considering doing business with had many consumer complaints about its service relative to the number of policies it sold.

### 5. Comfort
You should feel comfortable with your insurance purchase, whether you buy it from a local agent, directly from the company over the phone, or over the Internet. Make sure that the agent or company will be easy to reach if you have a question or need to file a claim.
