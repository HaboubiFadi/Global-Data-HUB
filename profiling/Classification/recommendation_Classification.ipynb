# I) Sampling

##                     




## 1) Undersampling

```python
print(Counter(y))

undersample = RandomUnderSampler()
undersampled_X, undersampled_y = undersample.fit_resample(X_scaled, y.to_numpy())
print(Counter(undersampled_y))

undersampled_X=pd.DataFrame(undersampled_X,columns=df.drop('recommandation', axis = 1).columns)
undersampled_y=pd.DataFrame(undersampled_y,columns=['recommandation'])
```

## 2) OverSampling

```python
Knn_nighbert_params(X_scaled,y,12)
```

```python

smote = SMOTE(sampling_strategy='auto',k_neighbors=1)
print(Counter(y))

oversampling_X, oversampling_y = smote.fit_resample(X,y)

print(Counter(oversampling_y))
oversampling_X=pd.DataFrame(oversampling_X,columns=df.drop('recommandation', axis = 1).columns)
oversampling_y=pd.DataFrame(oversampling_y,columns=['recommandation'])
```

## 3) Conclusion

* Given the stark imbalance in our dataset, with classes 0 and 1 dominating the landscape, comprising 41.1% and 26.1% respectively, while the remaining classes 2 and 3 lag behind at 17.6% and 15.1%, we opt for SMOTE oversampling. This choice ensures that we preserve the integrity of our data while mitigating the impact of class imbalance. Additionally, our dataset demonstrates strong performance with the KNN model, signifying that the introduction of synthetic samples through SMOTE won't introduce undue bias or inaccuracies. Thus, SMOTE emerges as the ideal approach to bolster our dataset and enhance the robustness of our models without compromising on the validity of our results.

## Add model/scaler into the table models in Postgres sql 

```python
import psycopg2
import pickle
db_config = {
    'dbname': 'postgres',
    'user': 'admin',
    'password': 'admin',
    'host': 'localhost',  
    'port': '5432'  
}



model_name="Recommendation_Classification_model"
scaler_name="Recommendation_Classification_scaler"
model_blob = pickle.dumps(SVC_model)
scaled_blob=pickle.dumps(scaled)
```

```python
try:
    # Connect to PostgreSQL
    conn = psycopg2.connect(**db_config)
    cursor = conn.cursor()

   
    # Insert the model into the table
    insert_model_query = """
    INSERT INTO model_storage (model_name, model_data)
    VALUES (%s, %s);
    """
    cursor.execute(insert_model_query, (model_name, model_blob))
    conn.commit()

    # Insert the scaler into the table
    insert_scaler_query = """
    INSERT INTO scalers_storage (scaled_name, scaled_data)
    VALUES (%s, %s);
    """
    cursor.execute(insert_scaler_query, (scaler_name, scaled_blob))
    conn.commit()
    print("scaler saved to PostgreSQL successfully!")

except Exception as e:
    print(f"An error occurred: {e}")
finally:
    # Close the cursor and connection
    if cursor:
        cursor.close()
    if conn:
        conn.close()
```
